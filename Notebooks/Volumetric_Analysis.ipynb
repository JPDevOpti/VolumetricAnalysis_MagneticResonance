```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo
plt.style.use('seaborn-v0_8')  # Usar una versión específica del estilo seaborn
sns.set_theme(style="whitegrid")  # Configurar el tema de seaborn

# Lista de las 16 regiones principales
main_regions = [
    'Left-Thalamus', 'Right-Thalamus',
    'Left-Hippocampus', 'Right-Hippocampus',
    'Left-Amygdala', 'Right-Amygdala',
    'Left-Accumbens-area', 'Right-Accumbens-area',
    'BrainSegVol', 'BrainSegVolNotVent',
    'lhCortexVol', 'rhCortexVol',
    'CortexVol', 'SubCortGrayVol',
    'TotalGrayVol', 'EstimatedTotalIntraCranialVol'
]

# Transforma el DataFrame a formato largo solo para las regiones principales
df_long = pd.melt(
    df_final,
    id_vars=['Biomarkers code', 'E280A group'],
    value_vars=main_regions,
    var_name='Brain Region',
    value_name='Volume'
)

# Crear figura con dos subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 16))

# Configuración común para ambos gráficos
for ax in [ax1, ax2]:
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.tick_params(axis='both', which='major', labelsize=10)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# Primer gráfico: primeras 8 regiones
sns.barplot(
    data=df_long[df_long['Brain Region'].isin(main_regions[:8])],
    x='Brain Region',
    y='Volume',
    hue='E280A group',
    ci='sd',
    ax=ax1,
    palette='Set2'
)
ax1.set_title('Comparison of Subcortical Brain Region Volumes by Group', 
              pad=20, fontsize=14, fontweight='bold')
ax1.set_xlabel('Brain Region', fontsize=12, labelpad=10)
ax1.set_ylabel('Volume (mm³)', fontsize=12, labelpad=10)
ax1.tick_params(axis='x', rotation=45)
ax1.legend(title='Group', title_fontsize=12, fontsize=10)

# Segundo gráfico: últimas 8 regiones
sns.barplot(
    data=df_long[df_long['Brain Region'].isin(main_regions[8:])],
    x='Brain Region',
    y='Volume',
    hue='E280A group',
    ci='sd',
    ax=ax2,
    palette='Set2'
)
ax2.set_title('Comparison of Global Brain Volume Measures by Group', 
              pad=20, fontsize=14, fontweight='bold')
ax2.set_xlabel('Brain Region', fontsize=12, labelpad=10)
ax2.set_ylabel('Volume (mm³)', fontsize=12, labelpad=10)
ax2.tick_params(axis='x', rotation=45)
ax2.legend(title='Group', title_fontsize=12, fontsize=10)

# Ajustar el diseño
plt.tight_layout()

# Añadir un título general
fig.suptitle('Brain Volume Analysis by Region and Group', 
             fontsize=16, fontweight='bold', y=0.95)

# Añadir una nota al pie
plt.figtext(0.99, 0.01, 
            'Note: Error bars represent standard deviation', 
            ha='right', fontsize=10, style='italic')

plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind, shapiro, mannwhitneyu

region_columns = df_final.columns[2:]
group_col = 'E280A group'
groups = df_final[group_col].unique()

results = []

for region in region_columns:
    data1 = df_final[df_final[group_col] == groups[0]][region].dropna()
    data2 = df_final[df_final[group_col] == groups[1]][region].dropna()
    shapiro_p1 = shapiro(data1).pvalue if len(data1) >= 3 else None
    shapiro_p2 = shapiro(data2).pvalue if len(data2) >= 3 else None
    both_normal = (
        shapiro_p1 is not None and shapiro_p2 is not None and
        shapiro_p1 > 0.05 and shapiro_p2 > 0.05
    )
    if both_normal:
        test = "t-test"
        stat, p_val = ttest_ind(data1, data2)
    else:
        test = "Mann-Whitney U"
        if len(data1) >= 3 and len(data2) >= 3:
            stat, p_val = mannwhitneyu(data1, data2, alternative='two-sided')
        else:
            p_val = None
    significant = p_val is not None and p_val < 0.05
    results.append({
        'Region': region,
        f'Shapiro p ({groups[0]})': shapiro_p1,
        f'Shapiro p ({groups[1]})': shapiro_p2,
        'Test used': test,
        'p-value': p_val,
        'Statistically significant': significant
    })

comparison_table = pd.DataFrame(results)
comparison_table.to_csv("brain_regions_group_comparison.csv", index=False)
print("Table saved as brain_regions_group_comparison.csv")
display(comparison_table)
```

```python
import pandas as pd
from scipy.stats import ttest_ind, shapiro, mannwhitneyu
import numpy as np

region_columns = df_final.columns[2:]
group_col = 'E280A group'
groups = df_final[group_col].unique()

results = []

for region in region_columns:
    data1 = df_final[df_final[group_col] == groups[0]][region].dropna()
    data2 = df_final[df_final[group_col] == groups[1]][region].dropna()
    shapiro_p1 = shapiro(data1).pvalue if len(data1) >= 3 else None
    shapiro_p2 = shapiro(data2).pvalue if len(data2) >= 3 else None
    both_normal = (
        shapiro_p1 is not None and shapiro_p2 is not None and
        shapiro_p1 > 0.05 and shapiro_p2 > 0.05
    )
    if both_normal:
        test = "Welch's t-test"
        stat, p_val = ttest_ind(data1, data2, equal_var=False)  # Welch's correction
    else:
        test = "Mann-Whitney U"
        if len(data1) >= 3 and len(data2) >= 3:
            stat, p_val = mannwhitneyu(data1, data2, alternative='two-sided')
        else:
            p_val = None
    
    # Calculate median for each group
    median1 = np.median(data1) if len(data1) > 0 else None
    median2 = np.median(data2) if len(data2) > 0 else None
    
    significant = p_val is not None and p_val < 0.05
    results.append({
        'Region': region,
        f'Shapiro p ({groups[0]})': shapiro_p1,
        f'Shapiro p ({groups[1]})': shapiro_p2,
        'Test used': test,
        'p-value': p_val,
        f'Median ({groups[0]})': median1,
        f'Median ({groups[1]})': median2,
        'Statistically significant': significant
    })

comparison_table = pd.DataFrame(results)
comparison_table.to_csv("brain_regions_group_comparison.csv", index=False)
print("Table saved as brain_regions_group_comparison.csv")
display(comparison_table)
```
