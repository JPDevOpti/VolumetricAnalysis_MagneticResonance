```python
import pandas as pd
from scipy.stats import ttest_ind, shapiro, mannwhitneyu

region_columns = df_final.columns[2:]
group_col = 'E280A group'
groups = df_final[group_col].unique()

results = []

for region in region_columns:
    data1 = df_final[df_final[group_col] == groups[0]][region].dropna()
    data2 = df_final[df_final[group_col] == groups[1]][region].dropna()
    shapiro_p1 = shapiro(data1).pvalue if len(data1) >= 3 else None
    shapiro_p2 = shapiro(data2).pvalue if len(data2) >= 3 else None
    both_normal = (
        shapiro_p1 is not None and shapiro_p2 is not None and
        shapiro_p1 > 0.05 and shapiro_p2 > 0.05
    )
    if both_normal:
        test = "t-test"
        stat, p_val = ttest_ind(data1, data2)
    else:
        test = "Mann-Whitney U"
        if len(data1) >= 3 and len(data2) >= 3:
            stat, p_val = mannwhitneyu(data1, data2, alternative='two-sided')
        else:
            p_val = None
    significant = p_val is not None and p_val < 0.05
    results.append({
        'Region': region,
        f'Shapiro p ({groups[0]})': shapiro_p1,
        f'Shapiro p ({groups[1]})': shapiro_p2,
        'Test used': test,
        'p-value': p_val,
        'Statistically significant': significant
    })

comparison_table = pd.DataFrame(results)
comparison_table.to_csv("brain_regions_group_comparison.csv", index=False)
print("Table saved as brain_regions_group_comparison.csv")
display(comparison_table)
```

```python
import pandas as pd
from scipy.stats import ttest_ind, shapiro, mannwhitneyu
import numpy as np

region_columns = df_final.columns[2:]
group_col = 'E280A group'
groups = df_final[group_col].unique()

results = []

for region in region_columns:
    data1 = df_final[df_final[group_col] == groups[0]][region].dropna()
    data2 = df_final[df_final[group_col] == groups[1]][region].dropna()
    shapiro_p1 = shapiro(data1).pvalue if len(data1) >= 3 else None
    shapiro_p2 = shapiro(data2).pvalue if len(data2) >= 3 else None
    both_normal = (
        shapiro_p1 is not None and shapiro_p2 is not None and
        shapiro_p1 > 0.05 and shapiro_p2 > 0.05
    )
    if both_normal:
        test = "Welch's t-test"
        stat, p_val = ttest_ind(data1, data2, equal_var=False)  # Welch's correction
    else:
        test = "Mann-Whitney U"
        if len(data1) >= 3 and len(data2) >= 3:
            stat, p_val = mannwhitneyu(data1, data2, alternative='two-sided')
        else:
            p_val = None
    
    # Calculate median for each group
    median1 = np.median(data1) if len(data1) > 0 else None
    median2 = np.median(data2) if len(data2) > 0 else None
    
    significant = p_val is not None and p_val < 0.05
    results.append({
        'Region': region,
        f'Shapiro p ({groups[0]})': shapiro_p1,
        f'Shapiro p ({groups[1]})': shapiro_p2,
        'Test used': test,
        'p-value': p_val,
        f'Median ({groups[0]})': median1,
        f'Median ({groups[1]})': median2,
        'Statistically significant': significant
    })

comparison_table = pd.DataFrame(results)
comparison_table.to_csv("brain_regions_group_comparison.csv", index=False)
print("Table saved as brain_regions_group_comparison.csv")
display(comparison_table)
```
