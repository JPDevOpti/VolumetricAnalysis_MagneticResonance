```python
import pandas as pd

# Reemplaza 'ruta/a/tu/archivo.xlsx' con la ruta real de tu archivo
ruta_archivo_excel = '/home/jp/Repositorios/VolumetricAnalysis_MagneticResonance/Data/brain_volumes.xlsx'

# Lee la hoja específica del archivo Excel
df_volumes = pd.read_excel(ruta_archivo_excel)
# Selecciona las primeras 60 filas (índices 0 a 59)
df_volumes = df_volumes[:61]
# Para verificar, puedes imprimir la forma (shape) del nuevo DataFrame
print(f"Dimensiones del DataFrame filtrado: {df_volumes.shape}")
df_volumes.sort_values(by='patient_id', inplace=True)
display(df_volumes)
```

```python
import pandas as pd

# Asumiendo que tus DataFrames ya están cargados:
# df_demography = pd.read_csv(...)  o similar
# df_volumes = pd.read_csv(...) o similar

# Combina los DataFrames usando las columnas de identificación de pacientes
df_combined = pd.merge(
    df_demography,
    df_volumes,
    left_on='Biomarkers code',        # Columna de ID en el DataFrame izquierdo (df_demography)
    right_on='patient_id', # Columna de ID en el DataFrame derecho (df_volumes)
    how='inner'            # Tipo de join, 'inner' mantiene solo los pacientes presentes en ambos DFs
                           # Puedes usar 'outer', 'left', o 'right' según necesites
)

# Opcional: Puedes eliminar una de las columnas de ID si ya no la necesitas
df_combined = df_combined.drop('patient_id', axis=1) # Elimina la columna redundante
df_combined = df_combined.drop('Code', axis=1) # Elimina la columna 
df_combined = df_combined.drop('Birthdate', axis=1) # Elimina la columna 
df_combined = df_combined.drop('Sex', axis=1) # Elimina la columna 
df_combined = df_combined.drop('MRI1 date', axis=1) # Elimina la columna 
df_combined = df_combined.drop('Age at MRI', axis=1) # Elimina la columna 


# Ahora df_combined contiene las columnas de ambos DataFrames
display(df_combined)
```

```python
import pandas as pd
from scipy.stats import ttest_ind, shapiro, mannwhitneyu

region_columns = df_combined.columns[2:]
group_col = 'E280A group'
groups = df_combined[group_col].unique()

results = []

for region in region_columns:
    data1 = df_combined[df_combined[group_col] == groups[0]][region].dropna()
    data2 = df_combined[df_combined[group_col] == groups[1]][region].dropna()
    shapiro_p1 = shapiro(data1).pvalue if len(data1) >= 3 else None
    shapiro_p2 = shapiro(data2).pvalue if len(data2) >= 3 else None
    both_normal = (
        shapiro_p1 is not None and shapiro_p2 is not None and
        shapiro_p1 > 0.05 and shapiro_p2 > 0.05
    )
    if both_normal:
        test = "t-test"
        stat, p_val = ttest_ind(data1, data2)
    else:
        test = "Mann-Whitney U"
        if len(data1) >= 3 and len(data2) >= 3:
            stat, p_val = mannwhitneyu(data1, data2, alternative='two-sided')
        else:
            p_val = None
    significant = p_val is not None and p_val < 0.01
    results.append({
        'Region': region,
        f'Shapiro p ({groups[0]})': shapiro_p1,
        f'Shapiro p ({groups[1]})': shapiro_p2,
        'Test used': test,
        'p-value': p_val,
        'Statistically significant': significant
    })

comparison_table = pd.DataFrame(results)
comparison_table.to_csv("brain_regions_group_comparison.csv", index=False)
print("Table saved as brain_regions_group_comparison.csv")
display(comparison_table)
```
