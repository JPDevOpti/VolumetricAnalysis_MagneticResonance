```python
import pandas as pd
from scipy import stats
from IPython.display import display # To display the table nicely in Jupyter

# Asumiendo que tu DataFrame se llama df_filtrado

# 1. Filtrar solo los grupos GK y GW
df_gw_gk = df_filtrado[df_filtrado['E280A group'].isin(['GK', 'GW'])].copy()

# Verificar si tenemos datos para ambos grupos
if 'GK' not in df_gw_gk['E280A group'].unique() or 'GW' not in df_gw_gk['E280A group'].unique():
    print("Error: No se encontraron datos para ambos grupos 'GK' y 'GW' en el DataFrame.")
# Verificar si el dataframe filtrado está vacío
elif df_gw_gk.empty:
    print("Error: El DataFrame filtrado para los grupos 'GK' y 'GW' está vacío.")
else:
    # --- Cálculos ---

    # a) Edad ('Age at MRI') - Media (Desviación Estándar)
    age_summary = df_gw_gk.groupby('E280A group')['Age at MRI'].agg(['mean', 'std', 'count'])
    # Asegurarse que ambos grupos existen en el summary
    if 'GK' in age_summary.index and 'GW' in age_summary.index:
        age_gk_str = f"{age_summary.loc['GK', 'mean']:.2f} ({age_summary.loc['GK', 'std']:.2f})"
        age_gw_str = f"{age_summary.loc['GW', 'mean']:.2f} ({age_summary.loc['GW', 'std']:.2f})"
        n_gk = int(age_summary.loc['GK', 'count'])
        n_gw = int(age_summary.loc['GW', 'count'])

        # Prueba T para comparar edad entre GK y GW
        gk_ages = df_gw_gk[df_gw_gk['E280A group'] == 'GK']['Age at MRI'].dropna()
        gw_ages = df_gw_gk[df_gw_gk['E280A group'] == 'GW']['Age at MRI'].dropna()
        if len(gk_ages) > 1 and len(gw_ages) > 1:
            t_stat_age, p_val_age = stats.ttest_ind(gk_ages, gw_ages)
            p_val_age_str = f"{p_val_age:.3f}"
            if p_val_age < 0.001: p_val_age_str = "<0.001"
        else:
            p_val_age_str = "N/A" # Not enough data
    else:
        print("Error: Falta uno o ambos grupos ('GK', 'GW') después de agrupar por edad.")
        age_gk_str, age_gw_str, n_gk, n_gw, p_val_age_str = "Error", "Error", 0, 0, "Error"


    # b) Sexo - N (%)
    # Verificar si hay datos de sexo antes de agrupar
    if not df_gw_gk[['E280A group', 'Sex']].dropna().empty:
        sex_counts = df_gw_gk.groupby('E280A group')['Sex'].value_counts().unstack(fill_value=0)
        # Asegurarse de que existan las columnas M y F incluso si no hay datos para un sexo
        if 'M' not in sex_counts.columns: sex_counts['M'] = 0
        if 'F' not in sex_counts.columns: sex_counts['F'] = 0
        sex_counts = sex_counts[['M', 'F']] # Asegurar orden

        sex_totals = sex_counts.sum(axis=1)
        # Evitar división por cero si un grupo no tiene datos de sexo
        sex_percentages = sex_counts.apply(lambda x: (x / sex_totals[x.name]) * 100 if sex_totals[x.name] > 0 else 0, axis=1)

        # Asegurarse que ambos grupos existen en los conteos
        male_gk_str = f"{sex_counts.loc['GK', 'M']} ({sex_percentages.loc['GK', 'M']:.1f})" if 'GK' in sex_counts.index else "N/A"
        female_gk_str = f"{sex_counts.loc['GK', 'F']} ({sex_percentages.loc['GK', 'F']:.1f})" if 'GK' in sex_counts.index else "N/A"
        male_gw_str = f"{sex_counts.loc['GW', 'M']} ({sex_percentages.loc['GW', 'M']:.1f})" if 'GW' in sex_counts.index else "N/A"
        female_gw_str = f"{sex_counts.loc['GW', 'F']} ({sex_percentages.loc['GW', 'F']:.1f})" if 'GW' in sex_counts.index else "N/A"

        # Prueba de Fisher (más adecuada para tablas 2x2, como Sexo vs Grupo)
        # Crear tabla de contingencia solo con datos no nulos
        contingency_table_sex = pd.crosstab(df_gw_gk['E280A group'], df_gw_gk['Sex'])
        
        # Verificar si la tabla es válida (2x2) para Fisher
        if contingency_table_sex.shape == (2, 2) and contingency_table_sex.notna().all().all():
            oddsr, p_val_sex = stats.fisher_exact(contingency_table_sex)
            p_val_sex_str = f"{p_val_sex:.3f}"
            # Si p-valor es muy pequeño, mostrar "<0.001"
            if p_val_sex < 0.001: p_val_sex_str = "<0.001"
        else:
             # Intentar Chi-cuadrado si Fisher no aplica (e.g., tabla no 2x2 pero con suficientes datos)
            try:
                chi2, p_val_sex_chi, dof, expected = stats.chi2_contingency(contingency_table_sex)
                p_val_sex_str = f"{p_val_sex_chi:.3f} (Chi2)"
                if p_val_sex_chi < 0.001: p_val_sex_str = "<0.001 (Chi2)"
            except ValueError: # Pocos datos para Chi2
                p_val_sex_str = "N/A" # No se puede calcular test

    else:
        print("Advertencia: No hay datos de sexo disponibles para calcular la distribución.")
        male_gk_str, female_gk_str, male_gw_str, female_gw_str, p_val_sex_str = "N/A", "N/A", "N/A", "N/A", "N/A"


    # --- Crear la tabla DataFrame ---
    # Solo crear tabla si n_gk y n_gw son válidos (mayores que 0)
    if n_gk > 0 and n_gw > 0 :
        table_data = {
            'Variable': [
                f'Age - Mean (SD)',
                'Sex - N (%)',
                '  Male',
                '  Female'
            ],
            f'GK (n={n_gk})': [
                age_gk_str,
                '', # Fila de encabezado para Sexo
                male_gk_str,
                female_gk_str
            ],
            f'GW (n={n_gw})': [
                age_gw_str,
                '', # Fila de encabezado para Sexo
                male_gw_str,
                female_gw_str
            ],
            'p-value (GK vs GW)': [
                p_val_age_str,
                '', # Fila de encabezado para Sexo
                '', # p-value general para Sexo se muestra en la siguiente línea
                p_val_sex_str
            ]
        }

        summary_table = pd.DataFrame(table_data)
        summary_table.set_index('Variable', inplace=True)

        # --- Mostrar la tabla ---
        print("\n--- Summary Table (GK vs GW) ---")
        display(summary_table) # Muestra la tabla formateada en el notebook

        print("\nNotes:")
        print("- Age is presented as Mean (Standard Deviation).")
        print("- Sex is presented as Count (Percentage).")
        print("- P-values compare GK vs GW groups using T-test for Age and Fisher's Exact Test (or Chi2) for Sex.")
        print("- Variables like Education, MMSE, CDR are not included as they were not found in the provided DataFrame.")
    else:
        print("\nNo se pudo generar la tabla resumen porque uno o ambos grupos (GK, GW) no tienen datos suficientes.")
```
